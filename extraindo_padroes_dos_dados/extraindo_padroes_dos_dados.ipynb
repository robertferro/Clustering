```python
from sklearn.preprocessing import Normalizer
```

```python
normalizer = Normalizer()

values = normalizer.fit_transform(dados.values)
```

```python
values
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
def melhor_numero_clusters(df,n1,n2):
  clusters = []
  scores = []
  for nc in range(n1, n2):
      kmeans = KMeans(n_clusters=nc)
      kmeans.fit(df)
      labels = kmeans.labels_
      ss=silhouette_score(df, labels,metric='euclidean')
      clusters.append(nc)
      scores.append(ss)
      melhor_score = max(scores)
      n = scores.index(melhor_score)
      numero = clusters[n]
  return (numero,melhor_score)
```

```python
melhor_numero_clusters(values,2,8)
```

```python
for nc in range(2, 11):
  print('Numero de Clusters', nc)
  kmeans = KMeans(n_clusters=nc)
  kmeans.fit(values)
  labels = kmeans.labels_
  ss=silhouette_score(values, labels)
  print('Silhouette Score', ss)
  print('='*80)
```

```python
kmeans = KMeans(n_clusters=6 , n_init=10, max_iter=300)
y_pred = kmeans.fit_predict(values)
```
